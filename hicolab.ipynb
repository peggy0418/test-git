```python
import matplotlib.pyplot as plt
import numpy as np

def draw_equilateral_triangle(ax,side_length, x_start, y_start):
    """
    使用 matplotlib.pyplot 绘制等边三角形
    """
    # 计算三角形的顶点坐标
    x = [x_start, x_start + side_length, x_start + side_length / 2, x_start]
    y = [y_start, y_start, y_start + side_length * np.sqrt(3) / 2, y_start]

    # 绘制三角形
    ax.plot(x, y)

def main():
    """
    主函数
    """
    # 创建子图
    fig, ax = plt.subplots()

    # 设置起始点坐标
    x_start, y_start = 0, 0

    # 绘制等边三角形
    for _ in range(3):
        draw_equilateral_triangle(ax, 100, x_start, y_start)
        x_start += 100  # 移动到下一个位置

    # 设置坐标轴范围

    # 可选：添加标题和标签
    ax.set_title("Turtle Graphics: Equilateral Triangle")
    ax.set_xlabel("X 坐标")
    ax.set_ylabel("Y 坐标")

    # 显示图形
    plt.show()

if __name__ == "__main__":
    main()
```
